```python
dataset = dataset.dropna(subset=['TOTAL_PEOPLE'])
dataset.isnull().sum()
```

```python
dataset['APPT_MADE_DATE'] = dataset['APPT_MADE_DATE'].str.split(' ').str[0]
dataset['APPT_MADE_DATE'] = pd.to_datetime(dataset['APPT_MADE_DATE'], format='%m/%d/%Y', errors='coerce')
dataset['Month'] = dataset['APPT_MADE_DATE'].dt.month
dataset = dataset.dropna(subset=['Month'])
dataset['Month'] = dataset['Month'].astype(int)
```

```python
counts = dataset["Month"].value_counts().sort_index()
ax = counts.plot(kind='bar', color='skyblue')
plt.xlabel('Month')
plt.ylabel('Count')
plt.title('Appointments Made Per Month')
for i, count in enumerate(counts):
    ax.text(i, count + 0.1, str(count), ha='center', va='bottom', fontsize=10)
plt.show()
```

```python
dataset["TOTAL_PEOPLE"] = dataset["TOTAL_PEOPLE"].replace("WIN",1).fillna(1).astype(int)
dataset["APPT_START_DATE"] = dataset["APPT_START_DATE"].astype(str)
dataset["APT_Month"] = dataset["APPT_START_DATE"].str.split('/').str[0]
dataset = dataset.dropna(subset=['APT_Month'])
dataset['APT_Month'] = dataset['APT_Month'].astype(int, copy=True)
```

```python
data = dataset.groupby("APT_Month")["TOTAL_PEOPLE"].sum().reset_index()
plt.scatter(data["APT_Month"], data["TOTAL_PEOPLE"])
plt.plot(data["APT_Month"], data["TOTAL_PEOPLE"])
plt.ticklabel_format(style='plain', axis='y')
plt.xticks(data['APT_Month'])
plt.ylabel('Total People')
plt.xlabel('Month')
plt.title('Total People For Each Month')
plt.grid()
plt.show()
```

```python
step_data = dataset[["APPT_START_DATE", "Month", "TOTAL_PEOPLE"]].copy() 
step_data["APPT_DAY"] = step_data["APPT_START_DATE"].str.split('/').str[1]
step_data = step_data.dropna(subset=['APPT_DAY'])
step_data['APPT_DAY'] = step_data['APPT_DAY'].astype(int, copy=True)
step_data=step_data[step_data['Month'] == 12]
step_data.drop(['APPT_START_DATE',"Month"], axis=1, inplace=True)
step_data = step_data.groupby("APPT_DAY")["TOTAL_PEOPLE"].sum().reset_index()
```

```python
plt.figure(figsize=(12, 6))
plt.step(step_data["APPT_DAY"], step_data["TOTAL_PEOPLE"])
plt.ticklabel_format(style='plain', axis='y')
plt.xlim((1,31))
plt.xticks(step_data["APPT_DAY"])
plt.ylabel('Total People')
plt.xlabel('Day')
plt.title('Total People For Each Day Of December')
plt.show()
```

```python
funnel_data = {
    "Labels" : ["Completed Appointments", "Canceled Appointments"],
    "Values" : [
                dataset["APPT_CANCEL_DATE"].isna().sum(),
                dataset["APPT_CANCEL_DATE"].notna().sum()
                ]
}
```

```python
explode = (0, 0.5)
fig, ax = plt.subplots()
ax.pie(funnel_data["Values"], labels=funnel_data["Labels"], autopct='%1.1f%%', explode=explode,shadow={'ox': -0.04, 'edgecolor': 'none', 'shade': 0.9})
plt.title("Breakdown of Appointment Types (Log10 Scale)")
plt.show()
```

```python
bar_w_time_data = dataset[["APPT_START_DATE", "Month", "TOTAL_PEOPLE"]].copy() 
bar_w_time_data["APPT_DAY"] = bar_w_time_data["APPT_START_DATE"].str.split('/').str[1]
bar_w_time_data = bar_w_time_data.dropna(subset=['APPT_DAY'])
bar_w_time_data['APPT_DAY'] = bar_w_time_data['APPT_DAY'].astype(int, copy=True)
bar_w_time_data.drop(['APPT_START_DATE'], axis=1, inplace=True)
bar_w_time_data = bar_w_time_data.groupby(["Month","APPT_DAY"])["TOTAL_PEOPLE"].sum().reset_index()
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
complete_ids = pd.Series(range(1, 32), name='APPT_DAY')
plt.ticklabel_format(style='plain', axis='y')
plt.xticks(bar_w_time_data['APPT_DAY'])
for i in range(1,13,1):
    filtered_data = bar_w_time_data[bar_w_time_data["Month"] == i].copy()
    filtered_data_sums = filtered_data.groupby('APPT_DAY')['TOTAL_PEOPLE'].sum().reset_index()
    filtered_data_sums = filtered_data_sums.sort_values(by='TOTAL_PEOPLE', ascending=True)
    filtered_data_sums = pd.merge(complete_ids.to_frame(), filtered_data_sums, on='APPT_DAY', how='left')
    filtered_data_sums['TOTAL_PEOPLE'] = filtered_data_sums['TOTAL_PEOPLE'].fillna(0)
    
    if i == 1:
        bottom = np.zeros(len(filtered_data_sums))
    plt.bar(filtered_data_sums['APPT_DAY'], filtered_data_sums['TOTAL_PEOPLE'], label=f'Month {i}', bottom=bottom, width=0.5)
    bottom += filtered_data_sums['TOTAL_PEOPLE'].values
ax.legend(title='Categories', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.ylabel("Count")
plt.xlabel("Day")
plt.title("Distribution of Attendees Per Month and Day")
plt.show()
```

```python
time_delta = dataset[["APPT_MADE_DATE", "APPT_START_DATE"]].copy()
time_delta["APPT_START_DATE"] = time_delta["APPT_START_DATE"].str.split(' ').str[0]
time_delta["APPT_START_DATE"] = pd.to_datetime(time_delta["APPT_START_DATE"], format='%m/%d/%Y', errors='coerce')
time_delta["APPT_MADE_DATE"] = time_delta["APPT_MADE_DATE"].astype(str).str.split(' ').str[0]
time_delta["APPT_MADE_DATE"] = pd.to_datetime(time_delta["APPT_MADE_DATE"], format='%Y-%m-%d', errors='coerce')
time_delta["TIME_DELTA"] = time_delta['APPT_START_DATE'] - time_delta['APPT_MADE_DATE']
time_delta["TIME_DELTA"] = time_delta["TIME_DELTA"].dt.days
time_delta_counts = time_delta["TIME_DELTA"].value_counts().sort_index().reset_index()
time_delta_counts["index"] = time_delta_counts["index"].astype(int)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(time_delta_counts["index"],time_delta_counts["TIME_DELTA"])
plt.xticks(time_delta_counts["index"])
plt.ylabel("Count")
plt.xlabel("Day Between Appointment Creation - Appointment Day")
plt.title("Day Difference between Appointment Creation - Appointment Day Counts")
plt.show()
```
